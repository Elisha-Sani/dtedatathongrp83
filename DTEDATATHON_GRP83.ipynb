```python
#importing the pandas library
import pandas as pd

#reading data that is uploaded in the colab database

new_excel_path = "/content/drive/My Drive/Urb&EneGdp.xlsx"

#Reading the excel data into a pandas dataframe named nf(new frame)
#format : 'dataframename = pd.read_excel(excel_declared_path)'

nf = pd.read_excel(new_excel_path)

#printing the data

print(nf)

```

```python
#Ploting a line plot for 'Year' against 'Urbanization rate (%)'
#importing libraries

import pandas as pd
import matplotlib.pyplot as plt

#Since we had already created a dataframe, we do not need to create another one
#Extracting data for the line plot from the dataframe
year = nf['Year']
urbanization_rate = nf['Urbanization rate (%)']

#Creating a line plot
plt.plot(year, urbanization_rate, marker='s', linestyle='-')

#Adding labels and title
plt.xlabel('Year')
plt.ylabel('Urbanization rate (%)')
plt.title('Urbanization Rate Over Time')

#Displaying the plot
#Adding a grid though it is an optional process
plt.grid()
plt.show()
```

```python
#Creating a line plot of Year against energy consumption
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt

#Extracting data from our dataframe
year = nf['Year']
ene_cons = nf['Energy consumption (TWh)']

#ploting
plt.plot(year, ene_cons, marker='s', linestyle='-', color='green')

#Adding Labels and Title
plt.xlabel('Year') #x-axis
plt.ylabel('Energy Consumption (TWh)') #y-axis
plt.title('Energy Consumption Over Time')

#Display the plot
plt.grid()
plt.show()
```

```python
#creating a dual-line plot for the 2 sets of data
#import
import pandas as pd
import matplotlib.pyplot as plt


#Extracting data from our dataframe
years = nf['Year']
urbanization_rate = nf['Urbanization rate (%)']
ene_cons = nf['Energy consumption (TWh)']



#Creating a figure(fig) & axis for the plot
fig, ax1 = plt.subplots()

#Ploting energy consumption on the left axis
ax1.set_xlabel('Year')
ax1.set_ylabel('Energy consumption (TWh)', color='tab:blue')
ax1.plot(years, ene_cons, color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

#Creating a second y-axis
ax2= ax1.twinx()
ax2.set_ylabel('Urbanization rate (%)', color='tab:red')
ax2.plot(years, urbanization_rate, color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Add a legend for both lines
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
ax1.legend(lines, labels, loc='upper left')

#adding a title
plt.title('Energy Consumption and Urbanization Over Time')

#Displaying the plot
plt.grid()
plt.show()
```

```python
#Creating a stacked plot
#import
import pandas as pd
import matplotlib.pyplot as plt

#Extracting data from our dataframe
years = nf['Year']
urbanization_rate = nf['Urbanization rate (%)']
ene_cons = nf['Energy consumption (TWh)']

#Creating a plot
plt.stackplot(years, urbanization_rate, ene_cons, labels=['Energy consumption (TWh)', 'Urbanization rate (%)'], alpha=0.7)

#Adding Labels and Buttons
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Energy Consumption and Urbanization Rate Over Time')

#legend
plt.legend(loc='upper left')

#Displaying the chart
plt.show()
```
